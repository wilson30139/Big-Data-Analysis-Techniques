```python
!pip install twstock 
!pip install yahoo_fin


from twstock import Stock
from yahoo_fin.stock_info import *

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.font_manager as fm





plt.rcParams['font.family'] = 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False
```

```python
dataList = pd.read_csv('./台積電股票2020資料.csv')
dataList
```

```python
stock_no = '2330.TW'

data = get_data(stock_no , start_date = '2021/02/01')
data.head()
```

```python
d2 = {'Date': data.index,
        'Close': data['close']
        }

newDf = pd.DataFrame(d2)
newDf
```

```python
"""
CSV
"""
dataList['Date'] = pd.to_datetime(dataList['Date'])
fig, ax = plt.subplots(figsize=(4,3),dpi=200)

plt.plot(dataList["Date"], dataList["Close"], color = "r", label = "台積電")
months = mdates.MonthLocator(interval=1, bymonthday=-1)
months_fmt = mdates.DateFormatter('%Y年-%m月')
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(months_fmt)
ax.legend(loc='upper right', ncol=2, fontsize=6)

ax.tick_params(axis='x', which='major', labelrotation=45)
plt.xticks(fontsize=6)
```

```python
"""
Yahoo API
"""
newDf['Date'] = pd.to_datetime(newDf['Date'])
fig, ax = plt.subplots(figsize=(4,3),dpi=150)
plt.plot(newDf["Date"], newDf["Close"], color = "b", label = "台積電(from Yahoo Stock API)")
months = mdates.MonthLocator(interval=1, bymonthday=-1)
months_fmt = mdates.DateFormatter('%Y年-%m月')
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(months_fmt)
ax.legend(loc='upper right', ncol=2, fontsize=6)
ax.tick_params(axis='x', which='major', labelrotation=45)
plt.xticks(fontsize=6)


plt.show()
```

```python
import numpy as np
import seaborn as sns
from matplotlib.pyplot import figure

"""
CSV
"""
df=pd.DataFrame(dataList["Close"].values,
                index=dataList["Date"],
                columns=['A'])

df.reset_index(inplace=True)
df.columns = ['Date','收市價']
df.head()

df['年月'] = df['Date'].dt.strftime('%Y年-%m月')
fig, ax = plt.subplots()
fig.set_size_inches((12,4))
sns.boxplot(x='年月',y='收市價',data=df,ax=ax)
ax.tick_params(axis='x', which='major', labelrotation=45)

```
